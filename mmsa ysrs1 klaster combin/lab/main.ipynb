```python
import pandas as pd
from combinator import ful_combinator
from compute_functionals import functional5_8
import matplotlib.pyplot as plt

df = pd.read_csv("input_data.csv", sep = ';', decimal = ',')
print(df)
```

```python
# case 1
comb = ful_combinator(list(range(0,12)), [6,6])
#print("получено " + str(len(comb)) + " вариантов")

result_data = pd.DataFrame(columns = ["разбиение", "значение ФКР"])

print(len(comb))

for i in comb:
    variant = str(i[0]) + '-' + str(i[1])
    result_data = result_data.append({"разбиение" : variant, "значение ФКР" : functional5_8([df.iloc[i[0]], df.iloc[i[1]]])}, ignore_index= True)
```

```python
# lets  save result to excel file
result_data.to_excel("result1.xlsx")
```

```python
min_fkr_index = result_data[['значение ФКР']].idxmin()

print("минимальное значение ФКР соответвует разбиению: ")
print(result_data.iloc[min_fkr_index])

fig = plt.figure(figsize = [10,20])
ax = fig.add_subplot(111, projection='3d')

np_cluster1 = df.iloc[comb[int(min_fkr_index)][0]].to_numpy()
np_cluster2 = df.iloc[comb[int(min_fkr_index)][1]].to_numpy()

ax.scatter(np_cluster1[:,0],np_cluster1[:,1],np_cluster1[:,2], s = 100, marker = "o")
ax.scatter(np_cluster2[:,0],np_cluster2[:,1],np_cluster2[:,2], s = 100, marker = ">")

plt.show()
```

```python
# case 2
comb = ful_combinator(list(range(0,12)), [4,4,4])
print("получено " + str(len(comb)) + " вариантов")

result_data_2 = pd.DataFrame(columns = ["разбиение", "значение ФКР"])

for i in comb:
    variant = str(i[0]) + '-' + str(i[1]) + '-' + str(i[2])
    result_data_2 = result_data_2.append({"разбиение" : variant, "значение ФКР" : functional5_8([df.iloc[i[0]], df.iloc[i[1]], df.iloc[i[2]]])}, ignore_index= True)
```

```python
# lets  save result to excel file
result_data_2.to_excel("result2.xlsx")
```

```python
min_fkr_index = result_data_2[['значение ФКР']].idxmin()

print("минимальное значение ФКР соответвует разбиению: ")
print(result_data_2.iloc[min_fkr_index])

fig = plt.figure(figsize = [10,20])
ax = fig.add_subplot(111, projection='3d')

np_cluster1 = df.iloc[comb[int(min_fkr_index)][0]].to_numpy()
np_cluster2 = df.iloc[comb[int(min_fkr_index)][1]].to_numpy()
np_cluster3 = df.iloc[comb[int(min_fkr_index)][2]].to_numpy()

ax.scatter(np_cluster1[:,0],np_cluster1[:,1],np_cluster1[:,2], s = 100, marker = "o")
ax.scatter(np_cluster2[:,0],np_cluster2[:,1],np_cluster2[:,2], s = 100, marker = ">")
ax.scatter(np_cluster3[:,0],np_cluster3[:,1],np_cluster3[:,2], s = 100, marker = "s")

plt.show()
```
